# Chocolate sales across the world

```python
# Libraries
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\dagui\Downloads\Chocolate Sales.csv")
df.head(15)
```

```python
# Cleansing

df["Amount"] = (
    df["Amount"]
        .str.replace("$","",regex = False)
        .str.replace(",","",regex = False)
        .astype(float)
)
df["Date"] = (
    pd.to_datetime(df["Date"], dayfirst = True, cache = True)
)
df.head()
```

```python
df.dtypes
```

```python
df["Day"] = df["Date"].dt.day
df["Month"] = df["Date"].dt.month
df["Year"] = df["Date"].dt.year
df.head()
```
