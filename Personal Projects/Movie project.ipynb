```python
import kagglehub
import pandas as pd

# Download latest version
path = kagglehub.dataset_download("chaitanyahivlekar/large-movie-dataset")
df = pd.read_csv(r"C:\Users\dagui\.cache\kagglehub\datasets\chaitanyahivlekar\large-movie-dataset\versions\1\movies_dataset.csv")
```

```python
del df['Unnamed: 0'], df['User_Id']
df_small = df.sample(n=5000) # Select 40 random rows of your dataset
df_small.head()
```

```python
df_small['Genre'].dropna()
df_small.head()
```

```python
# This separates the genres
df_small['Genre']  = df_small['Genre'].str.split('|')

from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
genres_encoded = mlb.fit_transform(df_small['Genre'])

genres_df = pd.DataFrame(genres_encoded, columns = mlb.classes_, index = df_small.index)
print(genres_df.head())
```

```python
df_final = pd.concat([df_small[['Movie_Name']], genres_df],axis=1)
```

```python
df_final.head()
```

```python
features_matrix = genres_df.values
print(features_matrix)
```

```python
import numpy as np


def cosine_sim(v1,v2):
    simil = np.dot(v1,v2)/(np.linalg.norm(v1)*np.linalg.norm(v2))
    return simil

print(f"similitud entre Gladiator y I'm not there: {cosine_sim(features_matrix[0],features_matrix[1])}")
```
