<h1 style="font-size: 38px; color: orange; text-align: center;">K-Means Clustering</h1>

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
df.head()
```

```python
X = df[['total_bill','size']]

model = KMeans(n_clusters = 6,max_iter = 300, random_state = True)

model.fit(X)
```

```python
model_predict = model.predict(X)
centroids = model.cluster_centers_
print(model.cluster_centers_)
```

```python

plt.figure(figsize=(7,5))
plt.scatter(X['total_bill'], X['size'], c = model_predict, s = 50, cmap = 'rainbow')
plt.scatter(centroids[:,0], centroids[:,1], c = 'black', s = 200, alpha = 1)
```
